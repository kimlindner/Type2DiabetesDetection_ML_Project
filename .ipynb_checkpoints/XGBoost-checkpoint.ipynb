XGBoost is one of the implementations of gradient boosted trees.

```python
import xgboost as xgb
from xgboost import plot_tree
import matplotlib.pyplot as plt
from ipynb.fs.full.Functions import load_train_test_data, load_train_test_oversampled, model_eval, run_gridsearchCV, run_randomizedsearchCV, load_model
```

### Loading Data

```python
X_train, X_test, y_train, y_test = load_train_test_data()
```

```python
X_train2, X_test2, y_train2, y_test2 = load_train_test_oversampled()
```

### Training with normal data (not oversampled)

- base model

```python
xgb_classifier = xgb.XGBClassifier(seed=0).fit(X_train,y_train)
```

```python
print(xgb_classifier)
```

```python
# Visualize
plot_tree(xgb_classifier)
plt.gcf().set_size_inches(18.5, 10.5)
```

```python
# evaluation
model_eval(xgb_classifier, X_train, X_test, y_train, y_test, confusion=True, roc=True)
```

#### Grid search with 10-fold cross validation

```python
grid_params = {
        'max_depth': [3, 4, 5, 6, 7, 8, 9, 10],
        'n_estimators': [60, 100, 140, 200, 260],
        'learning_rate': [0.3, 0.1, 0.01, 0.05, 0.001]
        }

run_gridsearchCV(xgb_classifier, grid_params, X_train, y_train)
```

```python
acc_best_model = load_model(xgb_classifier, 'accuracy')
model_eval(acc_best_model, X_train, X_test, y_train, y_test, confusion=True, roc=True)
```

```python
run_gridsearchCV(xgb_classifier, grid_params, X_train, y_train, scoring='recall')
```

```python
acc_best_model = load_model(xgb_classifier, 'recall')
model_eval(acc_best_model, X_train, X_test, y_train, y_test, confusion=True, roc=True)
```

```python
from scipy.stats import uniform as sp_randFloat
from scipy.stats import randint as sp_randInt

random_params = {
    'max_depth': sp_randInt(3, 10),
    'n_estimators': sp_randInt(50, 100),
    'learning_rate': sp_randFloat()
}
run_randomizedsearchCV(xgb_classifier, random_params, X_train, y_train)
```

```python
acc_best_model = load_model(xgb_classifier, 'accuracy', search='random')
model_eval(acc_best_model, X_train, X_test, y_train, y_test, confusion=True, roc=True)
```

```python
run_randomizedsearchCV(xgb_classifier, random_params, X_train, y_train, scoring='recall')
```

```python
acc_best_model = load_model(xgb_classifier, 'recall', search='random')
model_eval(acc_best_model, X_train, X_test, y_train, y_test, confusion=True, roc=True)
```

#### oversampled
