Now we can look at how the models compare against each other during the training rounds. The observed models are the best variations obtained during the experiments.

```python
def rolling_avg(x, n=3) :
    arr = np.cumsum(x, dtype=float)
    arr[n:] = arr[n:] - arr[:-n]
    return arr[n - 1:] / n


def plot_learning_curve(models,n=3):
    '''
    This function will plot the learning curve for the given model 
    
    '''
    # import packages needed
    import pandas as pd
    import matplotlib.pyplot as plt
    import numpy as np
    import sklearn
    from sklearn.metrics import f1_score

    
    custom_f1 = f1_score
    custom_f1.__dict__['average'] = 'macro'
    custom_f1.average = 'macro'
    
    # get data
    X_train, X_test, y_train, y_test = load_train_test_data(verbose=False)
    
    
    
    colors = ['green','blue','red','purple','black','yellow']
    steps = 40
    plt.figure(figsize=(10,5))
    scores = []
    for j,model in enumerate(models):
        model_score = []
        for i in np.linspace(.1,1,steps):
            new_model = sklearn.clone(model)
            sample_size = int(len(X_train)*i)
            new_model.fit(X_train.head(sample_size),y_train.head(sample_size))
            pred_test = new_model.predict(X_test)
            model_score.append(custom_f1(y_test,pred_test))
        scores.append((model,model_score))
        
    # plotting
        
    for j,model_score in enumerate(scores):
        plt.plot(np.linspace(.1,1,steps-n+1),rolling_avg(model_score[1],n=n), color=colors[j], 
                 linestyle='--',linewidth=1, markersize=1,label=type(model_score[0]).__name__)
    
    plt.xlabel('Training size')
    plt.ylabel('F1 score')
    plt.legend()
    plt.title('F1 x Training Size')
    return
```

```python
plot_learning_curve([forest,svm,LR,XGB,tree],n=5)
```

From the training process we can see an interesting information: both the SVM and the random forest managed to learn a good amount with very little data, however we also need to keep in mind that our test set is smewhat small, and flutuations in this graph could be an result of that.

<div style="border-bottom: 3px solid black"></div>

One other thing we could do is look at the features importance, but since while using an non-linear svm model the feature lose their meaning we will need to consider only the values for the random forest.

```python
importance = sorted(list(zip(X_train.columns,forest.feature_importances_)), key = lambda x: x[1],reverse=True)
feature,value = [a[0] for a in importance],[a[1] for a in importance]

plt.bar(feature,value, zorder=3)
plt.xticks(rotation=65, ha='right')
plt.xlabel('Feature')
plt.grid()
plt.ylabel('Importance')
plt.title('Feature importance')
plt.tight_layout()
plt.show()
```

As we can see from the values, the age and the usage of medicine has a deep impact on the model decision, however, the BPLevel is not as high as expected since during the data analysis there was an correlation of 0.4 , almost as high as the RegularMedicine.

## 3) Comparing models against external work

```python
our_models = [LR,svm,tree,forest]
model_names = [type(LR).__name__,type(svm).__name__,type(tree).__name__,type(forest).__name__]
ext_score = [0.890,0.906,0.891,0.959]
our_score = []

for model in our_models:
    v_test = model_eval(model, X_train, X_test, y_train, y_test,
                           acc=False, f1=True, recall=False, precision=False)[0][1]
    our_score.append(v_test)
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(111)

first_model = model_names[0]


for model,score in zip(model_names,our_score):
    if model == first_model:
        ax1.scatter(model,score,c='r',marker="o",label='Our model')
    else:
        ax1.scatter(model,score,c='r',marker="o")
        
for model,score in zip(model_names,ext_score):
    if model == first_model:
        ax1.scatter(model,score,c='b',marker="x",label='External model')
    else:
        ax1.scatter(model,score,c='b',marker="x")
    
plt.ylim(0,1)
plt.xticks(rotation=20)
plt.title('Model comparisson')
plt.ylabel("F1_score")
plt.legend()
plt.show()
```
